# Etape 3: Jointure
- La jointure est nécessaire pour lier les données entre eux. <br>
Ressources: https://bdnb.io/schema/latest/relationships.html

- La jointure a été faite par bloc car les colonnes primaires ne sont pas uniques pour les tables d'après l'analyse des duplicates. C'est pour ça il faut arriver à joindre les données par deux clés primaires ce qui va éviter les erreurs de jointure.

- Joindre batiment_construction avec rel_batiment_construction_adresse_unique suivant la clé primaire cle_interop_adr

```python
merged_data = pd.merge(adresse, rel_batiment_cons_adr_unique, on='cle_interop_adr', how='inner')
```

```python
#vérifier les duplicates:
duplicates_merged_data = merged_data[merged_data.duplicated(subset=['cle_interop_adr'], keep=False)]
print("Doublons dans adresse:")
print(duplicates_merged_data)
```

- Joindre batiment_construction avec rel_batiment_cons_adr_unique1 suivant la clé primaire batiment_construction_id

```python
merged_data_1 = pd.merge(batiment_construction,rel_batiment_cons_adr_unique1, on='batiment_construction_id', how='inner')
merged_data_1
#380590
```

- Join batiment_groupe_argiles and batiment_groupe to take  alea from batiment_groupe_argiles


```python
combined_data_adresse_batiment_rel = pd.merge(merged_data_1, merged_data, on=['batiment_construction_id','cle_interop_adr'], how='inner')
combined_data_adresse_batiment_rel
#185996
```

- Joindre batiment_groupe_rnc avec batiment_groupe suivant leur clé *primaire*

```python
merged_data3 = pd.merge(batiment_groupe_rnc, batiment_groupe, on='batiment_groupe_id', how='inner')
```

- Actuellement, la colonne annee_construction du csv batiment_groupe_rnc s'affiche sous forme (Exemple: [ "1780" ]), on va faire un nettoyage des données pour que la colonne soit mis sous forme année.


```python
def clean_value(x):
    if pd.notna(x):
        x = x.strip('[ "')
        x = x[0:4]
        return x
    else:
        return x
# Appliquer la fonction à la colonne 'l_annee_construction'
merged_data3['l_annee_construction'] = merged_data3['l_annee_construction'].apply(clean_value)
merged_data3['l_annee_construction'].unique()
```

#Jointure Gauche
- Les jointures gauches vont nous permettre d'ajouter de nouveaux caractéristiques sans perdre des valeurs de la dataset originale en mettant null quand les valeurs n'existent pas.

```python
merged_data_7 = pd.merge(combined_data_adresse_batiment_rel,batiment_groupe_argiles, on='batiment_groupe_id', how='left')
merged_data_8 = pd.merge(merged_data_7,merged_data3, on='batiment_groupe_id', how='left')
merged_data_9 = pd.merge(merged_data_8,batiment_groupe_bdtopo_bat, on='batiment_groupe_id', how='left')
merged_data_10 = pd.merge(merged_data_9,adresse_metrique, on='cle_interop_adr', how='left')
merged_data_10
```

# Visualisation des données

```python
# Importing necessary libraries
import pandas as pd          # For data manipulation and analysis
import numpy as np           # For numerical calculations
from shapely.geometry import Point  # For geometric operations
import pyproj               # For coordinate transformations
import folium               # For creating interactive maps
import geopandas as gpd     # For working with geospatial data
from folium.plugins import MarkerCluster  # For clustering markers on the map

# Defining Lambert-93 and WGS84 coordinate systems
lambert_93 = pyproj.Proj(init='epsg:2154')  # Lambert-93 coordinate reference system
wgs84 = pyproj.Proj(init='epsg:4326')       # WGS84 coordinate reference system (standard latitude and longitude)

# Reading CSV file containing address data into a pandas DataFrame
data = merged_data_10 # Update with your file path
# Conversion of Lambert-93 coordinates to WGS84 using vectorized operations
data['x_coords'] = data['WKT_x_y'].str.split().str[1].str[1:].astype(float)
data['y_coords'] = data['WKT_x_y'].str.split().str[2].str[:-1].astype(float)
lon, lat = pyproj.transform(lambert_93, wgs84, data['x_coords'], data['y_coords'])
data['geometry'] = [Point(lon[i], lat[i]) for i in range(len(lon))]

# Creating a GeoDataFrame from the DataFrame with a geometry column containing Point objects
gdf = gpd.GeoDataFrame(data, geometry='geometry')

# Creating a Folium map centered on the average latitude and longitude of the points with a zoom level of 7
mymap = folium.Map(location=[gdf['geometry'].y.mean(), gdf['geometry'].x.mean()], zoom_start=7)

# Creating a MarkerCluster object to group markers on the map
marker_cluster = MarkerCluster().add_to(mymap)

# Constructing the popup text for each row in the GeoDataFrame
gdf['popup_text'] = (
    "Latitude: " + gdf['geometry'].y.astype(str) + ", <br>" +
    "Longitude: " + gdf['geometry'].x.astype(str) + ", <br>" +
    "Adresse: " + gdf.apply(lambda row: f"{row['numero']} {row['type_voie']} {row['nom_voie']} {row['code_postal']}", axis=1) + ", <br>" +
    "Altitude_sol: " + gdf['altitude_sol'].astype(str) + ", <br>" +
    "Hauteur: " + gdf['hauteur'].astype(str) + ", <br>" +
    "Surface au sol[m²]: " + gdf['s_geom_cstr'].astype(str) + ", <br>" +
    "Aléa du risque argiles: " + gdf['alea'].astype(str) + ",<br>" +
    "Catégorie & Nature du bâtiment: " + gdf['l_nature'].astype(str) + ", <br>" +
    "Etat des bâtiments: " + gdf['l_etat'].astype(str) + ",  <br>" +
    "Usage principal du bâtiment: " + gdf['l_usage_1'].astype(str) + ", <br>" +
    "Usage secondaire du bâtiment: " + gdf['l_usage_2'].astype(str) + ", <br>" +
    "Nombre de bâtiments desservis par l'adresse: " + gdf['nb_bat_grp'].astype(str)
)

# Create markers with popup text using vectorized approach
for idx, row in gdf.iterrows():
    folium.Marker(location=[row['geometry'].y, row['geometry'].x], popup=row['popup_text']).add_to(marker_cluster)

# Saving the map to an HTML file
mymap.save('bdnp_17_data.html')
```
